#### Q1. Look at the data given below. Plot the data, find the outliers and find out  μ,σ,σ^2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as stat
import warnings
warnings.filterwarnings('ignore')
```

```python
name_of_company = ['Allied Signal','Bankers Trust','General Mills','ITT Industries','J.P.Morgan & Co.','Lehman Brothers',
      'Marriott','MCI','Merrill Lynch','Microsoft','Morgan Stanley','Sun Microsystems','Travelers','US Airways',
      'Warner-Lambert']
Measure_x = [24.23,25.53,25.41,24.14,29.62,28.25,25.81,24.39,40.26,32.95,91.36,25.99,39.42,26.71,35.00]
```

```python
# boxplot
plt.boxplot(Measure_x)
plt.show()
```

```python
# barplot
color_palette = "Set1"
sns.barplot(x=name_of_company,y=Measure_x,palette=color_palette)
plt.xticks(rotation=90)
plt.show()
```

```python
# Outliers
Measure_x = pd.Series(Measure_x)
q1 = np.quantile(Measure_x,0.25)
q3 = np.quantile(Measure_x,0.75)
q2 = np.median(Measure_x)
IQR = q3 - q1
upper_bond = q3+(1.5*IQR)
lower_bond = q1-(1.5*IQR)
outliers = Measure_x[(Measure_x <= lower_bond) | (Measure_x >= upper_bond)]
print('The Outlier in the boxplot:',outliers)
```

```python
print('Mean :',stat.mean(Measure_x))
print('Variance :',stat.variance(Measure_x))
print('Standard Deviation :',stat.stdev(Measure_x))
```
