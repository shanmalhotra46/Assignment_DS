```python
train1 = train.iloc[:,0:13]
train1 = pd.get_dummies(train1)
train1
```

```python
train2 = pd.concat([train1, train['Salary']],axis=1)
train2
```

```python
test1 = test.iloc[:,0:13]
test1 = pd.get_dummies(test1)
test1
```

```python
test2 = pd.concat([test1, test['Salary']],axis=1)
test2
```

```python
X = train2.values[:,0:102]
Y = train2.values[:,102]

x = test2.values[:,0:102]
y = test2.values[:,102]
```

Multinomial Naive Bayes

```python
classifier_mb = MB()

classifier_mb.fit(X,Y)
train_pred = classifier_mb.predict(X)
accuracy_train = np.mean(train_pred==Y)

test_pred = classifier_mb.predict(x)
accuracy_test = np.mean(test_pred==y)
print('Training accuracy is:',accuracy_train,'\n','Testing accuracy is:',accuracy_test)
```

**Gaussian Naive Bayes** 

```python
classifier_gb = GB()
classifier_gb.fit(X,Y) 
train_pred_g = classifier_gb.predict(X)
accuracy_train_g = np.mean(train_pred_g==Y)

test_pred_g = classifier_gb.predict(x)
accuracy_test_g = np.mean(test_pred_g==y)

print('Training accuracy is:',accuracy_train_g,'\n','Testing accuracy is:',accuracy_test_g)
```
